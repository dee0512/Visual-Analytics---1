#### Importing the dataset:

```python
import pandas
from bokeh.plotting import figure, output_notebook, show
import math

dataframe = pandas.read_excel(io='Homicide age adjusted indicator LIVE -05 20100919.xlsx',sheetname='Data',index_col=0,na_values='NA')

countries = list(dataframe.index)
```

#### For my first graph, I will be creating a scatter plot for the murder rate for all countries in 2005 : 


```python
p = figure(x_range = countries, x_axis_label='Countries', y_axis_label='Murder per 100,000, age adjusted', plot_width=3000)
p.xaxis.major_label_orientation = math.pi/2
p.circle(countries, dataframe[2005], size=10, color="navy", alpha=0.5)

output_notebook()
show(p)
```
